```python
import sympy as sym
import numpy as np
from IPython.display import Math, Latex, Markdown, display
import math
```

```python
w, nu, n, E , h, rho, A = sym.symbols('\\omega, \\nu, n, E , h, \\rho, A')
display(w, nu, n, E , h, rho, A)
```

```python
E = 140*10**9
rho = 3700
nu = 0.33
a = 2
b = 1
h = 0.01
```

```python
D = E*h**3/(12*(1-nu**2))
display(D)

def omega_mn(m,n):
    omega_mn = (np.pi)**2*((m/a)**2+(n/b)**2)*sym.sqrt(D/(rho*h))
    return omega_mn
    
check_m = 7
check_n = 7

wij = []
# Loop over pairs (m, n)
for i in range(1, check_m):  # Loop for m from 1 to 3
    for j in range(1, check_n):  # Loop for n from 1 to 3
        w = omega_mn(i, j)
        wij.append((w, i, j))  # Store (omega_mn, i, j) as a tuple

# Sort the list of tuples based on omega_mn
sorted_wij = sorted(wij)

# Display the sorted values and corresponding indices
for w, i, j in sorted_wij:
    ij = f"{i}{j}"
    display(Math(r'\omega_{{{}}} = {}'.format(ij, w)))

for w, i, j in sorted_wij:
    ij = f"{i}{j}"
    f = w/(2*np.pi)
    display(Math(r'f_{{{}}} = {}'.format(ij, f)))


# test = (np.pi)**2*((3/a)**2+(1/b)**2)*sym.sqrt(D/(rho*h))
# display(test)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define parameters
m = 1
n = 1

# Define mode shape function
def modeshape(x, y):
    return np.sin((m * np.pi * x) / a) * np.sin((n * np.pi * y) / b)

# Generate grid points
x = np.linspace(0, a, 100)
y = np.linspace(0, b, 50)
X, Y = np.meshgrid(x, y)

# Calculate mode shape
Z = modeshape(X, Y)

# Create smooth contour plot
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
contour = plt.contourf(X, Y, Z, cmap='hsv', levels=100)
plt.colorbar(contour, label='Mode Shape')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Smooth Contour Plot (m={}, n={})'.format(m, n))
plt.axis('equal')

# Create smooth 3D surface plot
ax = plt.subplot(1, 2, 2, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='hsv', rstride=1, cstride=1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('Mode Shape')
ax.set_title('Smooth 3D Surface Plot (m={}, n={})'.format(m, n))
ax.set_box_aspect([2, 1, 1])  # Set aspect ratio for 3D plot
plt.show()
```
